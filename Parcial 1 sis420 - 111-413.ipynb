```python
"""
El juego contiene a cuatro personajes: un pastor, una cabra, un lobo y una lechuga
Donde el pastor debe cruzar el rio con estos personajes, pero solo puede llevar
un personaje a la vez, y si deja a la cabra con la lechuga o al lobo con la cabra
estos se comen y pierde el juego.
El objetivo es llevar a todos los personajes a la otra orilla del rio sin que se
coman entre ellos.
El modelo de datos es una tupla de dos tuplas, donde la primera tupla contiene
los personajes en la orilla izquierda y la segunda tupla contiene los personajes
en la orilla derecha.
El algoritmo de busqueda es BFS (Breadth First Search) o busqueda en amplitud
segun el cual se expande el nodo mas cercano al nodo raiz. Se utiliza una cola 
como estructura de datos para almacenar los nodos a expandir. 
Se escogio este algoritmo de busqueda ya que es el mas adecuado para este problema 
debido a que se debe encontrar la solucion mas rapida posible. 

El agente de este problema es el pastor, ya que es el unico personaje que puede 
cruzar el rio y llevar a los otros personajes a la otra orilla.

El estado inicial es el siguiente:
((pastor,lobo,cabra,lechuga),(vacio))
El estado final es el siguiente:
((vacio),(pastor,lobo,cabra,lechuga))

El costo de cada movimiento es de 1, ya que solo se puede llevar un personaje a la
vez y se debe llevar al pastor para que no se coman los otros personajes.
"""
def camino(visited,node):
    #la funcion camino retorna la solucion del problema
  camino_solucion = [node]
  while camino_solucion[-1]["parent"]:
    camino_solucion.append(visited[tuple(camino_solucion[-1]["parent"])])
  return camino_solucion

def goal_state(game_state):
  #la funcion goal_state retorna true si el estado es el estado final y false si no lo es
  if game_state[0]:
    return False
  return True

def posible_movimiento(game_node):
    #la funcion posible_movimiento retorna una lista con los posibles movimientos, game_node es el estado actual
    #y moves es la lista de posibles movimientos
  moves = []
  inicio = list(game_node[0])
  meta = list(game_node[1])
  if "pastor" in inicio:
    inicio_copy = inicio[:] 
    meta_copy = meta[:]
    inicio_copy.remove("pastor")
    meta_copy.append("pastor")
    for move in inicio_copy:
      inicio_copy1 = inicio_copy[:]
      meta_copy2 = meta_copy[:]
      inicio_copy1.remove(move)
      meta_copy2.append(move)
      moves.append([tuple(inicio_copy1),tuple(meta_copy2)])
  else:
    inicio_copy = inicio[:]
    meta_copy = meta[:]
    meta_copy.remove("pastor")
    inicio_copy.append("pastor")
    for move in meta_copy:
      inicio_copy1 = inicio_copy[:]
      meta_copy2 = meta_copy[:]
      meta_copy2.remove(move)
      inicio_copy1.append(move)
      moves.append([tuple(inicio_copy1),tuple(meta_copy2)])
    moves.append([tuple(inicio_copy),tuple(meta_copy)])
  return moves

def condiciones(game_node):
    #la funcion condiciones retorna true si las condiciones del problema se cumplen y false si no se cumplen, las condiciones son:
    #si el pastor esta solo en la orilla izquierda y la cabra y la lechuga estan en la orilla derecha, el pastor no puede cruzar
    #si el pastor esta solo en la orilla derecha y la cabra y la lechuga estan en la orilla izquierda, el pastor no puede cruzar
    #si el pastor esta solo en la orilla izquierda y el lobo y la cabra estan en la orilla derecha, el pastor no puede cruzar
    #si el pastor esta solo en la orilla derecha y el lobo y la cabra estan en la orilla izquierda, el pastor no puede cruzar

  
  if "cabra" in game_node[0] and "lechuga" in game_node[0]: 
    return "pastor" in game_node[0]
  if "lobo" in game_node[0] and "cabra" in game_node[0]:
    return "pastor" in game_node[0]
  if "cabra" in game_node[1] and "lechuga" in game_node[1]:
    return "pastor" in game_node[1]
  if "lobo" in game_node[1] and "cabra" in game_node[1]:
    return "pastor" in game_node[1]
  return True

def search(visited,borde):
    #la funcion search retorna la solucion del problema y los nodos expandidos y visitados 
    #en borde se almacenan los nodos expandidos y en visited se almacenan los nodos visitados
    
  while borde:
    nodo = borde[0]
    borde = borde[1:]
    visited[tuple(nodo["state"])] = nodo
    if goal_state((nodo["state"])):
      return camino(visited,nodo)
    children = []
    for child_state in posible_movimiento(nodo["state"]):
      if condiciones(child_state):
        children.append(
          {
            "state" : tuple(child_state),
            "parent" : nodo["state"]
          }
          )
    for child in children:
      if child["state"] not in visited:
        borde.append(child)

inicio_1 = ( "pastor", "lobo", "cabra", "lechuga")
meta_2 = ()
root_node={
    "state" : [ inicio_1, meta_2],
    "parent" : None
  }
visited = {}
borde = [root_node]
camino_final = search(visited,borde)
solution = reversed(camino_final)
for i in solution:
  print(*i["state"])
```
